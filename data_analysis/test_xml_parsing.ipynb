```python
import pandas as pd 
import xml.etree.ElementTree as et 

xtree = et.parse("28-ex_xml_atu.xml")
xroot = xtree.getroot() 


# https://medium.com/@robertopreste/from-xml-to-pandas-dataframes-9292980b1c1c
```

```python
# https://data.gov.ua/dataset/a2d6c060-e7e6-4471-ac67-42cfa1742a19/resource/e21a1e57-051c-46ea-9c8e-8f30de7d863d
```

```python
col = ['OBL_NAME', 'REGION_NAME', 'CITY_NAME', 'CITY_REGION_NAME', 'STREET_NAME'] # RECORD

rows = []

for node in xroot: 
    
    s_name = node.attrib.get("name")
    
    s_mail = node.find("REGION_NAME").text if node is not None else None
    s_grade = node.find("CITY_NAME").text if node is not None else None
    s_age = node.find("CITY_REGION_NAME").text if node is not None else None
    street_name = node.find("STREET_NAME").text if node is not None else None
    
    rows.append({'REGION_NAME': s_mail, 'CITY_NAME': s_grade, 'CITY_REGION_NAME':s_age, \
                'STREET_NAME':street_name})
```

```python
out_df = pd.DataFrame(rows, columns = col)

out_df.info()
```
