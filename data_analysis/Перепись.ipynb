```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('03_021.csv')
# 3.2.1. Вікові коефіцієнти народжуваності, за рік
# http://database.ukrcensus.gov.ua/Mult/Dialog/statfile1_c_files/pasport.files/pasport/00_uk.htm#01

avg_wage = pd.read_csv('year_avg_wage_15-20.csv', delimiter='\t')
```

```python
plt.plot(data.year, data.loc[:, '15-49'])
plt.title('Среднее количество рожденных детей женщинами с 15 до 49 лет в 1990 до 2020')
plt.show()
```

```python
new = pd.DataFrame(data)
```

```python
plt.figure(figsize=(20,10))

plt.suptitle('количество рожденных детей и возраст женшины по годам 1990 - 2020', fontsize=20)

plt.plot(new.year, new[['15-49', ' 15-19', ' 20-24', ' 25-29', ' 30-34', ' 35-39', \
       ' 40-44', ' 45-49']])

plt.legend(['15-49', ' 15-19', ' 20-24', ' 25-29', ' 30-34', ' 35-39', \
       ' 40-44', ' 45-49'])

plt.show()
```

```python
tempo = new[26:]

plt.figure(figsize=(20,10))
plt.plot(tempo.year, tempo[['15-49']])

plt.legend(['15-49'])


plt.suptitle('количество рожденных детей, в среднем по годам с 2015 по 2019', fontsize=20)
plt.plot()

avg_wage = avg_wage[:5]
avg_wage

tempo['wage'] = avg_wage.avg_wage.values
```
